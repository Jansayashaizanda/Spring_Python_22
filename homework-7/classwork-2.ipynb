```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.special
```

```python
#задание 1.1
xx = np.linspace(-1,1,250)
yy = xx**2 * np.sin(1 / xx ** 2) + xx
fig, ax = plt.subplots()
ax.plot(xx, yy, label = 'x^2 sin(1 / x^2)')
ax.set_xlabel('x')
ax.set_xlabel('y')
ax.legend()
```

```python
#задание 1.2
xx = np.linspace(5,25,250)
yy = abs(1 - (1 / (1 + xx**2)) / (1 / (xx**2)))
fig, ax = plt.subplots()
ax.semilogy(xx, yy, label = 'x^2 sin(1 / x^2)')
ax.set_xlabel('x')
ax.set_xlabel('y')
ax.legend()
```

```python
#задание 1.3
xx = np.linspace(-10,10,100)
yy1 = scipy.special.gamma(xx)
yy2 = scipy.special.wofz(xx)
yy3 = scipy.special.lambertw(xx)
fig, ax = plt.subplots(1, 2)
ax[0].plot(xx, yy1, label = 'gamma function')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].legend()
ax[1].plot(xx, yy2, color = 'red', ls = '-.', label = 'Faddeeva function')
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')
ax[1].legend()
```

```python
#задание 2
xx = np.linspace(-6 * np.pi, 6 * np.pi, 100)
yy = np.sin(xx) / xx
fig, ax = plt.subplots(figsize=(10,2))
ax.plot(xx, yy, label = ('sin x / x'))
ax.set_xlabel('x')
ax.set_xlabel('y')
plt.xticks(np.linspace(-6 * np.pi, 6 * np.pi, 25), 
           [r'$-6\pi$', r'$-\frac{11\pi}{2}$', r'$-5\pi$', r'$-\frac{9\pi}{2}$', r'$-4\pi$', r'$-\frac{7\pi}{2}$', r'$-3\pi$', 
            r'$-\frac{5\pi}{2}$', r'$-2\pi$', r'$-\frac{3\pi}{2}$', r'$-\pi$', r'$-\frac{\pi}{2}$', r'$0$', r'$\frac{\pi}{2}$',
            r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$', r'$\frac{5\pi}{2}$', r'$3\pi$', r'$\frac{7\pi}{2}$', r'$4\pi$', 
            r'$\frac{9\pi}{2}$', r'$5\pi$', r'$\frac{11\pi}{2}$', r'$-6\pi$'])
ax.grid()
ax.legend()
```

```python
#задание 3.1
xx = np.linspace(-np.pi, np.pi, 100)
yy = np.linspace(-np.pi, np.pi, 100)
XX, YY = np.meshgrid(xx, yy)
fig, ax = plt.subplots()
ax.contourf(XX , YY, np.sin(XX) * np.sin(YY))
```
